```python
names = ["张三", "欧阳锋", "司马相如", "李雷", "东方不败"]
vip_list = []
# 目标：筛选名字字数 > 3 的人

for name in names:
    if len(str(name)) > 3:
        # 这里想把名字加进新列表
        vip_list.append(name) 

print(f"VIP客户名单有：{vip_list}")
```
